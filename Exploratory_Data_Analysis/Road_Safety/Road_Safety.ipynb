```python
def replace_values(df_main, df_replacement, column_name, column_name_desc):
  df_reference = df_metadata[df_metadata['field_name'] == column_name].loc[:,['code','label']]
  df_reference.columns = [column_name, column_name_desc]
  df_reference.set_index(column_name, inplace=True)
  dictionary = df_reference.to_dict(orient='dict')[column_name_desc]
  df_temp = df_main.copy()
  df_temp[column_name] = df_temp[column_name].astype('str')
  df_temp[column_name] = df_temp[column_name].replace(dictionary)
  return df_temp
```

```python
from matplotlib import pyplot as plt

def generate_count_plot(df_data, column_name, chart_title, xlabel, ylabel):
  # plot type
  sns.set(style="whitegrid")
  plt.figure(figsize=(8, 6))
  # generate the count plot for the variable
  sns.countplot(data=df_data, x=column_name, palette='viridis')
  # set the axis labels and title
  plt.title(chart_title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.xticks(rotation=90)
  # show the plot
  plt.show()

def generate_count_plot_horizontal(df_data, column_name, chart_title, xlabel, ylabel):
  # plot type
  sns.set(style="whitegrid")
  plt.figure(figsize=(8, 6))
  # generate the count plot for the variable
  sns.countplot(data=df_data, y=column_name, palette='viridis')
  # set the axis labels and title
  plt.title(chart_title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  #plt.xticks(rotation=90)
  # show the plot
  plt.show()

def generate_count_plot_with_hue(df_data, column_name, hue_column, chart_title, xlabel, ylabel, hue_label):
  # plot type
  sns.set(style="whitegrid")
  plt.figure(figsize=(8, 6))
  # generate the count plot for the variable
  ax = sns.countplot(data=df_data, x=column_name, hue=hue_column, palette='Set2')

  # Add legend title
  legend = ax.legend(title=hue_label)
  legend.get_title().set_fontsize('12')  # Set the font size if neede

  # set the axis labels and title
  plt.title(chart_title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.xticks(rotation=90)
  # show the plot
  plt.show()

def generate_count_plot_with_hue_horizontal(df_data, column_name, hue_column, chart_title, xlabel, ylabel, hue_label):
  # plot type
  sns.set(style="whitegrid")
  plt.figure(figsize=(8, 6))
  # generate the count plot for the variable
  ax = sns.countplot(data=df_data, y=column_name, hue=hue_column, palette='Set2')

  # Add legend title
  legend = ax.legend(title=hue_label)
  legend.get_title().set_fontsize('12')  # Set the font size if neede

  # set the axis labels and title
  plt.title(chart_title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  # show the plot
  plt.show()
```

```python
def generate_vehicle_specific_charts(df_main):
  df_copy = replace_values(df_main, df_metadata, 'road_type', 'road_type_desc')
  generate_count_plot(df_copy, 'road_type', 'Collisions by Road Type', 'Road Type','Collisions')

  df_copy = replace_values(df_main, df_metadata, 'first_road_class', 'first_road_class_desc')
  generate_count_plot(df_copy, 'first_road_class', 'Collisions by Road Class', 'Road Class','Collisions')

  df_copy = replace_values(df_main, df_metadata, 'junction_detail', 'junction_detail_desc')
  generate_count_plot(df_copy, 'junction_detail', 'Collisions by Junction Info', 'Junction Info','Collisions')

  df_copy = replace_values(df_main, df_metadata, 'speed_limit', 'speed_limit_desc')
  generate_count_plot(df_copy, 'speed_limit', 'Collisions by Speed limit of road', 'Speed Limit','Collisions')

  df_copy = df_main[df_main['pedestrian_location'] != 0]
  df_copy = replace_values(df_copy, df_metadata, 'pedestrian_location', 'pedestrian_location_desc')
  generate_count_plot(df_copy, 'pedestrian_location', 'Collisions by Pedestrian location', 'Pedestrian Location','Collisions')

  df_copy = df_main[df_main['pedestrian_movement'] != 0]
  df_copy = replace_values(df_copy, df_metadata, 'pedestrian_movement', 'pedestrian_movement_desc')
  generate_count_plot(df_copy, 'pedestrian_movement', 'Collisions by Pedestrian Movement', 'Pedestrian Movement','Collisions')
```
